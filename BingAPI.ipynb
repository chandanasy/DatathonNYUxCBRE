```python
def fetch_satellite_image(lat, lon, api_key, zoom_level=18, size="500,500"):
    url = f"https://dev.virtualearth.net/REST/v1/Imagery/Map/Aerial/{lat},{lon}/"
    params = {
        "mapSize": size,
        "zoomLevel": zoom_level,
        "key": api_key,
        "format": "jpeg"
    }
    response = requests.get(url, params=params)
    if response.status_code == 200:
        image_data = response.content  # Gets the image data
        file_name = f"{folder_path}/image_{lat}_{lon}.jpeg"  # Save in the Google Drive folder
        with open(file_name, 'wb') as f:
            f.write(image_data)  # Save the image as a file
        print(f"Image saved to {file_name}")
        return file_name
    else:
        print(f"Error {response.status_code}: {response.text}")  # This will print out the error code and message
        return None
```

```python
def process_coordinates(data, api_key):
    for index, row in data.iterrows():
        image_file = fetch_satellite_image(row['Latitude'], row['Longitude'], api_key)
        if image_file:
            print(f"Saved image to {image_file}")
        else:
            print(f"Failed to retrieve image for coordinates: {row['Latitude']}, {row['Longitude']}")

# Assuming 'data' is a pandas DataFrame with the columns 'Latitude' and 'Longitude'
api_key = "Ag6O9X8QcASByUEchrDiNAeriYzUW9QWODFir5tCTIIsPpMshfK1YNSvwQwvSLz7"  # Replace with your actual API key
process_coordinates(data, api_key)
```
